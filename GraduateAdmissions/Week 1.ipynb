## Looking at the data

Before we do any sort of modeling, it helps a great deal to do some exploratory data analysis. Why?
- Oftentimes, as data scientists, we're explaining our findings to a non-technical audience. We need to be able to tell a story with our data such that anybody can understand it.
- When we understand our data better, we can choose where exactly to focus our modeling.

```python
# import necessary modules
import pandas as pd
df = pd.read_csv('Admission_Predict_Ver1.1.csv')
df
```

This is how our data looks like in a Pandas Dataframe. Now I'll do some example exploratory data analysis using a Python library called Seaborn. There's a good tutorial on Seaborn right here:

## What kind of impact does research have on the probability of being admitted to grad school?

```python
# documentation for this function:
# https://seaborn.pydata.org/generated/seaborn.barplot.html
import seaborn as sns
ax = sns.barplot(x="Research", y="Chance of Admit ", data=df,ci=None)
```

Note that a '0' represents students without research experience, and '1' represents students with research experience. We can see that those with research experience have a higher chance of admit (how would we gauge if this difference is statistically significant? - think back to basic stats).
